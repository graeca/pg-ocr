```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

numValues1=5
numValues2=5
numValues=numValues1+numValues2






x = np.random.randint(25,100,numValues1)

y = np.random.randint(175,255,numValues2)
z = np.hstack((x,y))

z = z.reshape((numValues,1))
print(z.shape)
z = np.float32(z)
plt.figure()
plt.hist(z,256,[0,256]),plt.show()


# Define criteria = ( type, max_iter = 10 , epsilon = 1.0 )
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

# Set flags (Just to avoid line break in the code)
flags = cv2.KMEANS_RANDOM_CENTERS

# Apply KMeans
compactness,labels,centers = cv2.kmeans(z,2,None,criteria,10,flags)


A = z[labels==0]
B = z[labels==1]
print(labels)

# Now plot 'A' in red, 'B' in blue, 'centers' in yellow
plt.hist(A,256,[0,256],color = 'r')
plt.hist(B,256,[0,256],color = 'b')
plt.hist(centers,32,[0,256],color = 'y')
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

X = np.random.randint(25,50,(25,2))
Y = np.random.randint(60,85,(25,2))
Z = np.vstack((X,Y))

# convert to np.float32
Z = np.float32(Z)
print(Z.shape)
# define criteria and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(Z,2,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)

# Now separate the data, Note the flatten()
A = Z[label.ravel()==0]
B = Z[label.ravel()==1]
print(A)
# Plot the data
plt.scatter(A[:,0],A[:,1])
plt.scatter(B[:,0],B[:,1],c = 'r')
plt.scatter(center[:,0],center[:,1],s = 80,c = 'y', marker = 's')
plt.xlabel('Height'),plt.ylabel('Weight')
plt.show()
```

```python

import matplotlib.pyplot as plt
import cv2
from scipy import ndimage
from sklearn import cluster
#image = ndimage.imread("tmp1/test.png")
image=cv2.imread("tmp1/test2.png")
plt.figure(figsize = (15,8))
plt.imshow(image)
print(image.shape)
x, y, z = image.shape

image_2d = image.reshape(x*y, z)
print(image_2d.shape)

kmeans_cluster = cluster.KMeans(n_clusters=7)
print("--------")
kmeans_cluster.fit(image_2d)
print("=====")
cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_


plt.figure(figsize = (15,8))
plt.imshow(cluster_centers[cluster_labels].reshape(x, y, z))


```
