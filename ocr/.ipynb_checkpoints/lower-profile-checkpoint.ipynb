```python
import numpy as np
import cv2 as cv2
from matplotlib import pyplot as plt
from scipy.stats.stats import pearsonr 
def expSmooth(series,alfa):
    result=[series[0]]
    for n in range(1,len(series)):
        result.append(alfa*series[n]+(1-alfa)*result[n-1])
    return result
    


def findLowerProfile(img3):
    mtrx=[]
    for col in img3.T:
        #print(col)
        count=0
        for row in col:
            
            #print(row)
            if row==0:
                count=count+1
                #print(count)
            else: 
                break
                
    
        #print('count=',count)
        mtrx.append(count)
    mtrx=np.asarray(mtrx)
    h=img3.shape[1]
    mtrx=h-mtrx
    mtrx = mtrx/max(mtrx)
    #mtrx = mtrx/max(mtrx)
    return mtrx

imageSource = 'word/oti.png'
imageSource = 'word/paulos.png'

imageSource = 'test/0185.pdf300/308-1122-132-35.png'#PNP
imageSource = 'test/0185.pdf300/580-1726-132-35.png'#PNP


imageSource = 'test/0185.pdf300/1217-1486-120-38.png'#PA
imageSource = 'test/0162.pdf300/1490-173-131-38.png'#PNP
img = cv2.imread(imageSource,0)
 
# copy image to display all 4 variations
 
h_img = cv2.flip( img, 1 )
v_img = cv2.flip( h_img, -1 )
v_img= cv2.bitwise_not(v_img).copy()
x=0
Y=findLowerProfile(v_img )
plt.figure()
plt.xlabel('metr')
plt.imshow(v_img)

plt.figure()
plt.plot(Y)
alfa=0.7
Ysm=expSmooth(Y,alfa)
#plt.figure()
plt.plot(Ysm)
coef=pearsonr(Y,Ysm)
print(coef )





plt.figure()
#Yhist, bin_edges = np.histogram(Y, density=False)
Yhist, bin_edges = np.histogram(Y, density=True)

#plt.hist(Yhist, bins='auto')
plt.plot(Yhist)
print('bin_edges=',bin_edges)




Ycum=np.cumsum(Y)
x=np.arange(0,len(Ycum))
print(len(Ycum))
print(len(x))
plt.figure()
plt.plot(x,Ycum)

m = (len(x) * np.sum(x*Ycum) - np.sum(x) * np.sum(Ycum))/(len(x)*np.sum(x*x) - np.sum(x) ** 2)
b = (np.sum(Ycum) - m *np.sum(x)) / len(x)
print('m=',m)
print('b=',b)



```
