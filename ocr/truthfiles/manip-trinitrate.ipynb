```python
%%bash
IFS='
'
#∆αυῒδ
#Ῥωμαίοις
#ὑποστάσεως
words='∆αυ.δ
Ῥωμα.οις
.ποστ.σεως
πίστεως
ἅγιον
Κορινθίοις
φῶς
Ἰησοῦ
Χριστοῦ
Παῦλος
πατρὸς
υἱὸς
θεὸν
κύριος
πνεύματος
πνεῦμα
θεὸς
ὅτι
γὰρ
καὶ'

word='κύριος'

for word in `echo "$words"`
do

x=0
files=`find  pages -type f|sort`
for file in `echo "$files"`
do
#echo $file

cat "$file"|tr " " "\n"|sort|uniq -c|sed -rn 's@('$word')(.*)@\1\2@gp'

num=`cat "$file"|tr " " "\n"|sort|sed -rn 's@('$word')(.*)@\1\2@gp'|wc -l`
#echo $num

x=`expr $x + $num`
done
echo "number of occurence of ... $word: $x"
done


```
