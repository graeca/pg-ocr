```python

import matplotlib.pyplot as plt 
import numpy as np
import cv2
from scipy import interpolate

from scipy import optimize
from scipy.signal import argrelextrema

%run PgFunctions.py 

def hHist(img):
    newX=256
    newY=32
    img = cv2.bitwise_not(img)
    img = cv2.resize(img,(int(newX),int(newY)))
    img=img/255
    y=img.sum(axis=0)
    ymax=max(y)
    y=y/ymax
    #index=np.where(y == 0)
    #print(index)
    #print(y.shape)
    num=np.size(img,1)
    x = np.arange(num)
    #print(x.shape)
    fig, ax = plt.subplots()
    ax.plot(x,y)
    plt.show()
    return y
    
def vHist(img):
    newX=256
    newY=32
    img = cv2.bitwise_not(img)
    img = cv2.resize(img,(int(newX),int(newY)))
    img=img/255
    y=img.sum(axis=1)
    ymax=max(y)
    y=y/ymax
    
    print(max(y))
    
    
    
    #index=np.where(y == 0)
    #print(index)
    #print(y.shape)
    num=np.size(img,0)
    x = np.arange(num)
    #print(x.shape)
    fig, ax = plt.subplots()
    ax.plot(x,y)
    plt.show()
    return y
       

 


def drawMiddleLine(img2):
    yv=vHist(img2)

    #yv = np.array([1,2,3,4,5,4,3,2,1,2,3,2,1,2,3,4,5,6,5,4,3,2,1])
    # determine the indices of the local maxima
    maxInd = argrelextrema(yv, np.greater)
    # get the actual values using these indices
    ymax = yv[maxInd]  # array([5, 3, 6])
    #print(ymax)

    maxPoints=np.sort(ymax, axis=0)[::-1]
    print(maxPoints)

    m1=maxPoints[0]
    m2=maxPoints[1]
    Xind1=np.where(yv == m1)
    Xind2=np.where(yv == m2)
    x1=Xind1[0][0]
    x2=Xind2[0][0]
    xm=(x1+x2)//2
    #print(maxInd[0])
    #print(len(maxInd[0]))
    print(x1,x2,xm)
    w=img2.shape[1]
    lineThickness=3
    img2=cv2.line(img2, (5, xm), (w-5, xm), (0,255,0), lineThickness)
    
    return img2

def createPadding(img, margin):
    w=img.shape[1]
    #print(w)
    for i in range(10):
        img = np.insert(img, w, values=255, axis=1)

    for i in range(10):
        img = np.insert(img, 0, values=255, axis=1)    
    
    h=img.shape[0]
    #print(w)
    for i in range(10):
        img = np.insert(img, h, values=255, axis=0)
    
    for i in range(10):
        img = np.insert(img, 0, values=255, axis=0)
    
    
    return img


def findBiggestContour(cnt2):
    arealist=[1]
    for i in range(1,len(cnt2)):
        area=cv2.contourArea(cnt2[i])
        arealist.append(area)
        
        print("area=",area)
        
    #print(max(arealist))
    maxarea=max(arealist)
    for i in range(1,len(cnt2)):
        area=cv2.contourArea(cnt2[i])
        if maxarea==area:
            maxindex=i
            
            return  maxindex,maxarea


def applyCircleErosion(word_img):
    word_img=createPadding(word_img, 20) 
    kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    erode_img = cv2.erode(word_img, kernel, iterations=1)
    return erode_img
##############################
img2 = cv2.imread("word/pisteos.png",0);
img2 = cv2.imread("word/pneuma.png",0);
img2 = cv2.imread("test/0185.pdf300/1217-1486-120-38.png",0);
img2=applyCircleErosion(img2)
plt.figure()
plt.imshow(img2,cmap="gray")
newX=256
newY=32
img2 = cv2.resize(img2,(int(newX),int(newY)))
img2=drawMiddleLine(img2)
newX=256
newY=64
img2 = cv2.resize(img2,(int(newX),int(newY)))
img2=createPadding(img2, 10)
_, cnt1, _ = cv2.findContours(img2, cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE) 
maxindex1,_=findBiggestContour(cnt1)


x,y=img2.shape
word1Empty=np.zeros(img2.shape)+255
#contours_word1=cv2.drawContours(word1, cnt1, -1, (0,255,0), 1)
conimg2=cv2.drawContours(word1Empty, cnt1[maxindex1], -1, (0,255,0), 1)

plt.figure()
plt.imshow(conimg2,cmap="gray")
print("len=",len(cnt1))

count=0
#files = [line.rstrip() for line in open("word/pisteos.png.txt", 'r')]
files = [line.rstrip() for line in open("word/pneuma.png.txt", 'r')]

for file in files:
    print(file)
    count=count+1
    if count >9:
        continue
    img2 = cv2.imread(file,0);
    img2=applyCircleErosion(img2)
    newX=256
    newY=32

    img2 = cv2.resize(img2,(int(newX),int(newY)))
    img2=drawMiddleLine(img2)
    
    
    newX=256
    newY=64
    img2 = cv2.resize(img2,(int(newX),int(newY)))
    img2=createPadding(img2, 10)
    _, cnt2, _ = cv2.findContours(img2, cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
    
    
   #+++
    maxindex2,_=findBiggestContour(cnt2)
    
    x,y=img2.shape
    word1Empty=np.zeros(img2.shape)+255
    #contours_word1=cv2.drawContours(word1, cnt1, -1, (0,255,0), 1)
    conimg2=cv2.drawContours(word1Empty, cnt2[maxindex2], -1, (0,255,0), 1)
    
    
    plt.figure()
    plt.imshow(conimg2,cmap="gray")
    

    print("len=",len(cnt2))
    sd = cv2.createShapeContextDistanceExtractor() 
    try:
        abc = sd.computeDistance(cnt1[maxindex1],cnt2[maxindex2])
        print('abc=',abc)
    except:
        print("error")
    #print('abc=',abc)



```
